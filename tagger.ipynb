# Regex PoS Tagger



```python
import re
```

```python
with open ('morphodita-processed.txt', 'r') as txt:
    txt = txt.read()
```

```python
txt
```

```python
txt = re.sub('\t.*\t', ' ', txt)
```

```python
print(txt)
```

```python
txt = txt.split('\n')
```

```python
txt
```

```python
tagged = []
for tok in txt:
    if len(tok) > 0:
        tok = tok.replace(tok[-16:], '('+tok[-15]+')')
        tagged.append(tok)
```

```python
print(tagged)
```

```python
with open('princ_tagged.txt', 'x') as princ:
    for tok in tagged:
        princ.write(tok+' ')
```

```python
with open('princ_tagged.txt', 'r') as princ:
    princ = princ.read()
```
